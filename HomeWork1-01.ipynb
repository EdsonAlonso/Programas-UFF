> # (Item A) 
> A equação de Archie nos permite calcular a saturação de água em uma formação a partir das medidas de resistividade obtidas em perfis de poço. Para uma determinada profundidade, calcule a saturação de água:

> ## Equação de Archie: $S_w^2 = \frac{a ~ \times ~ R_w}{\phi^m ~ \times ~ R_t}$,
> em que $S_w ^ 2$ é o quadrado da saturação de água, $R_t$ é a resistividade da água na formação em $\Omega m$, $\phi$ é a porosidade $a$ é o coeficiente litológico. 

| Variavel | Valor |
|----------|-------|
| coef litológico ($a$) | 1 |
| expoente ($m$) | 2 |
| Porosidade ($\phi$) | 20 $\%$ |
| Resistividade da água ($R_w$) | 1.5 $ \Omega m$ |
| Resistividade medida ($R_t$) | 400.0 $\Omega m $ |

```python
# CODIGO AQUI !
a=1 #coef litoógico
m=2 #expoente
o=0.2 #porosidade
rw=1.5 #resistividade da agua
rt=400.0 #resistividade medida

x=( (a*rw)/((o**m)*rt) )
X=x**(1/2)
print(f'O nível de saturação de água é {X:.2f}')
```

> # (Item B)
> ### Cálculo do perfil de Saturação de água em profundidade: <br>
>Usando a equação de Archie e os valores de coeficiente litológico, porosidade e a resistividade da água apresentados na tabela acima, calcule o perfil de saturação de água em profundidade. <br>
> Construa os vetores prof (profundidade) usando o comando linspace do numpy e Rt (resistividade) através da função uniform do pacote random com base nas informações da tabela abaixo. Lembre-se que a amostragem indica o numero de medidas em profundidade, ou seja, para este exercício, são 200 observações entre o intervalo de profundidades estipulado. Apresente os gráficos em subplot dos perfis de resistividade e saturação de água.

| Variavel | topo($m$) | base($m$) | Amostragem |
|----------|-------|---|------------|
| profundidade ($m$) | 50.0 | 1000.0| 200 |
| resistividade ($\Omega m $)| 390.0| 410.0 | 200 |


```python
# Importe os Pacotes necessarios para que tudo funcione bem !
import numpy as np   #------------Necessario para criar o vetor prof------------------#
import matplotlib.pyplot as plt  #--------------Necessario para plotar----------------#
from random import uniform  #---------------Necessario para criar Rt-----------------#
```

```python
# Definicao das profundidades atraves do comando linspace
prof = np.linspace(50.0,1000.0,200,endpoint='True')
```

```python
# Definição das resistividades medidas através do comando uniform (estude o comando uniform para usa-lo adequadamente)
Rt = np.zeros(200)
for i in range(200):
    Rt[i] = uniform(390.0,410.0)
```

## <center> Cálculo da saturação de água via modelo de Archie: <center>


```python
# preencha as variaveis abaixo com os valores da tabela de dados:
a   = 1 # coef litologico
m   = 2 # expoente
phi = 0.2 # porosidade
Rw  =  1.5# resistividade da agua

# calculo via equacao de Archie:
S = np.sqrt( (a*Rw)/((o**m)*Rt) )
```

## <center> Log plots 

```python
# -------------- Subplot 1 (saturacao de agua) -------------------------------------------
plt.figure(figsize=(10,12)) # tamanho da janela de plot em pixels
ax1 = plt.subplot(121) # comando para subplot (121 = 1 linha, 2 colunas, posicao 1 do plot)
plt.plot(S,prof,'k',label = '$S_w$')
plt.title('Water Saturation log')
plt.ylabel('Depth (m)')
plt.xlabel('Water Saturation')
plt.legend()

# comando para inverter o eixo y e coloca-lo positivo para baixo
plt.gca().invert_yaxis()
# comando para colocar listras na janela de plot
plt.grid()
# comando para visualizar as janelas de plot

#----------------------SUBPLOT (resistividade)-------------------------------------------------------------------
ax2=plt.subplot(122, sharey=ax1) # comando para subplot (122 = 1 linha, 2 colunas, posicao 2 do plot)
plt.plot(Rt,prof,'r',label = '$R_t$')
plt.title('Resistivity log')
plt.xlabel('Resistivity ($\Omega$)')
plt.legend()

# comando para inverter o eixo y e coloca-lo positivo para baixo
plt.gca().invert_yaxis()
# comando para colocar listras na janela de plot
plt.grid()
# comando para visualizar as janelas de plot
plt.show()
```
